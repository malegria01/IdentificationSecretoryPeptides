## **Protocol 001**: Control Signal Peptide Extraction

**Input:** GFF file with annotations of *E. coli* proteome

**Output:** List of Uniprot IDs with experimental validation involved in cell secretion with Sec-pathway

```python
f = open("EColi Proteome GFF file.gff", "r")   #input gff file
x = f.readlines()

listas = []

for i in x:
    if "0000269" in i:      #Genbank keyword ECO: 0000269
        if "Signal" in i:
            if "Tat" not in i:
                listas.append(i)
```

```python
listas
```

```python
print("List of Uniprot IDS")
for i in listas:
    row = i.split("\t")
    print(row[0] +"["+row[3]+"-"+row[4]+"]")
```
